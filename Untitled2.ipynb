```python
x=df.drop(columns="House_Price")
y=df["House_Price"]
```

```python
x
```

```python
# As numeric variables have different numeric scales we use Scikit-learn library to fix that(Normalization of data)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
scaler.fit(x)
```

```python
# Note in this step we call on the data in x to be represented in the table format by using the x[:] and normalized
```

```python
x[:]= scaler.transform(x)
```

```python
x
```

```python
# The data is split in a training set for training the model and test set for testing the model created
```

```python
from sklearn.model_selection import train_test_split
x_train ,x_test,y_train ,y_test = train_test_split(x, y, test_size=0.2, random_state=20)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
model.fit(x_train, y_train)
```

```python
# y_pred are the House_Prices that are able to be predicted by the model created
```

```python
y_pred= model.predict(x_test)
```

```python
# These are the actual House_Price by y_test and y_pred are the House_Prices that were predicted by the nodel
```

```python
y_test[:10]
```

```python
y_pred[:10]
```

```python
# Determining the error to see how well the model has performed
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
# We now do some optimization to reduce the error
```

```python
from sklearn.linear_model import Lasso
lasso_model =Lasso(alpha=50, max_iter=100, tol=0.1)
lasso_model.fit(x_train, y_train)
```

```python
lasso_pred= lasso_model.predict(x_test)
```

```python
mean_absolute_error(y_test, lasso_pred)
```

```python
from sklearn.linear_model import Ridge
ridge_model= Ridge(alpha=50, max_iter=100, tol=0.1)
ridge_model.fit(x_train, y_train)
```

```python
ridge_pred= lasso_model.predict(x_test)
```

```python
mean_absolute_error(y_test, ridge_pred)
```
